```python
# Importing all the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pickle
```

```python
# Read the pickle file
with open('./Data/final_df.pkl', 'rb') as f:
    sales_data = pickle.load(f)
```

```python
# Convert 'date_time_local' to datetime format
sales_data['date_time_local'] = pd.to_datetime(sales_data['date_time_local'])

# Extract year and month from 'date_time_local'
sales_data['year'] = sales_data['date_time_local'].dt.year
sales_data['month'] = sales_data['date_time_local'].dt.month
```

```python
sales_data = sales_data[sales_data['year'].isin([2023, 2024])]
```

```python
# Split 'all_employees' column into individual employees in case of multiple employees per transaction
# We will then "explode" the data frame to have one employee per row
sales_data['all_employees'] = sales_data['all_employees'].str.split(',')

# Explode the employees column to separate multiple employees per transaction
sales_data_exploded = sales_data.explode('all_employees')
```

```python
# Group by employee, year, and month, and calculate total sales (subtotal) for each
employee_sales_by_month = sales_data_exploded.groupby(
    ['all_employees', 'year', 'month']
)['subtotal'].sum().reset_index()

# Sort by employee and date for better readability
employee_sales_by_month = employee_sales_by_month.sort_values(by=['all_employees', 'year', 'month'])

employee_sales_by_month.head()
```

```python
# Set the plot style for seaborn
sns.set(style="whitegrid")

# Create a pivot table to visualize data more effectively
pivot_table = employee_sales_by_month.pivot_table(
    values='subtotal', 
    index=['year', 'month'], 
    columns='all_employees', 
    aggfunc='sum'
)

# Create a figure for the plot
plt.figure(figsize=(14, 10))

# Plot a line plot for each employee's monthly sales
pivot_table.plot(kind='line', linewidth=2, marker='o', ax=plt.gca())

# Add titles and labels
plt.title('Total Sales by Each Employee per Month', fontsize=16)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Total Sales (Subtotal)', fontsize=12)

# Customize x-axis labels to show year and month
plt.xticks(ticks=range(len(pivot_table.index)), 
           labels=[f'{year}-{month:02d}' for year, month in pivot_table.index], 
           rotation=45, ha='right')

# Add a legend to identify each employee
plt.legend(title='Employee', bbox_to_anchor=(1.05, 1), loc='upper right')

plt.savefig('per_month_sales.png')

# Show the plot
plt.tight_layout()
plt.show()

```

```python
# Filter data for Kristin and Tianna
kristin_data = sales_data_exploded[sales_data_exploded['all_employees'] == 'Kristin Reid']
tianna_data = sales_data_exploded[sales_data_exploded['all_employees'] == 'Tianna Parks']

# Calculate key metrics
def calculate_metrics(data):
    total_sales = data['sold_at_price'].sum()
    num_transactions = len(data)
    avg_transaction = total_sales / num_transactions
    product_mix = data['classification_x'].value_counts()
    return total_sales, num_transactions, avg_transaction, product_mix

kristin_metrics = calculate_metrics(kristin_data)
tianna_metrics = calculate_metrics(tianna_data)

# Print results
print("Kristin Reid:")
print(f"Total sales: ${kristin_metrics[0]:.2f}")
print(f"Number of transactions: {kristin_metrics[1]}")
print(f"Average transaction value: ${kristin_metrics[2]:.2f}")
print("\nTianna Parks:")
print(f"Total sales: ${tianna_metrics[0]:.2f}")
print(f"Number of transactions: {tianna_metrics[1]}")
print(f"Average transaction value: ${tianna_metrics[2]:.2f}")

# Plotting
plt.figure(figsize=(15, 10))

# Sales comparison
plt.subplot(2, 2, 1)
sales_comparison = [kristin_metrics[0], tianna_metrics[0]]
plt.bar(['Kristin', 'Tianna'], sales_comparison)
plt.title('Total Sales Comparison')
plt.ylabel('Total Sales ($)')

# Transaction count comparison
plt.subplot(2, 2, 2)
transaction_comparison = [kristin_metrics[1], tianna_metrics[1]]
plt.bar(['Kristin', 'Tianna'], transaction_comparison)
plt.title('Number of Transactions Comparison')
plt.ylabel('Number of Transactions')

# Average transaction value comparison
plt.subplot(2, 2, 3)
avg_transaction_comparison = [kristin_metrics[2], tianna_metrics[2]]
plt.bar(['Kristin', 'Tianna'], avg_transaction_comparison)
plt.title('Average Transaction Value Comparison')
plt.ylabel('Average Transaction Value ($)')

# Product mix comparison
plt.subplot(2, 2, 4)
kristin_product_mix = kristin_metrics[3]
tianna_product_mix = tianna_metrics[3]
all_categories = list(set(kristin_product_mix.index) | set(tianna_product_mix.index))
kristin_values = [kristin_product_mix.get(cat, 0) for cat in all_categories]
tianna_values = [tianna_product_mix.get(cat, 0) for cat in all_categories]

x = range(len(all_categories))
width = 0.35
plt.bar([i - width/2 for i in x], kristin_values, width, label='Kristin')
plt.bar([i + width/2 for i in x], tianna_values, width, label='Tianna')
plt.xlabel('Product Categories')
plt.ylabel('Number of Transactions')
plt.title('Product Mix Comparison')
plt.xticks(x, all_categories, rotation=45, ha='right')
plt.legend()

plt.tight_layout()
plt.show()

# Sales over time
plt.figure(figsize=(12, 6))
kristin_daily_sales = kristin_data.groupby('date_time_local')['sold_at_price'].sum()
tianna_daily_sales = tianna_data.groupby('date_time_local')['sold_at_price'].sum()

plt.plot(kristin_daily_sales.index, kristin_daily_sales.values, label='Kristin')
plt.plot(tianna_daily_sales.index, tianna_daily_sales.values, label='Tianna')
plt.title('Daily Sales Comparison')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from datetime import timedelta

# Convert date_time_local to datetime
sales_data_exploded['date_time_local'] = pd.to_datetime(sales_data_exploded['date_time_local'])

# Filter data for Kristin and Tianna
kristin_data = sales_data_exploded[sales_data_exploded['all_employees'] == 'Kristin Reid']
tianna_data = sales_data_exploded[sales_data_exploded['all_employees'] == 'Tianna Parks']

# Find Tianna's start date
tianna_start_date = tianna_data['date_time_local'].min()

# Filter Kristin's data to match Tianna's time period
kristin_data_matched = kristin_data[kristin_data['date_time_local'] >= tianna_start_date]

# Calculate the end date (same number of days for both)
end_date = tianna_start_date + timedelta(days=(tianna_data['date_time_local'].max() - tianna_start_date).days)

# Apply the end date filter to both datasets
kristin_data_matched = kristin_data_matched[kristin_data_matched['date_time_local'] <= end_date]
tianna_data_matched = tianna_data[tianna_data['date_time_local'] <= end_date]

# Calculate key metrics
def calculate_metrics(data):
    total_sales = data['sold_at_price'].sum()
    num_transactions = len(data)
    avg_transaction = total_sales / num_transactions
    product_mix = data['classification_x'].value_counts()
    return total_sales, num_transactions, avg_transaction, product_mix

kristin_metrics = calculate_metrics(kristin_data_matched)
tianna_metrics = calculate_metrics(tianna_data_matched)

# Print results
print(f"Analysis period: {tianna_start_date.date()} to {end_date.date()}")
print("\nKristin Reid:")
print(f"Total sales: ${kristin_metrics[0]:.2f}")
print(f"Number of transactions: {kristin_metrics[1]}")
print(f"Average transaction value: ${kristin_metrics[2]:.2f}")
print("\nTianna Parks:")
print(f"Total sales: ${tianna_metrics[0]:.2f}")
print(f"Number of transactions: {tianna_metrics[1]}")
print(f"Average transaction value: ${tianna_metrics[2]:.2f}")

# Plotting
plt.figure(figsize=(15, 10))

# Sales comparison
plt.subplot(2, 2, 1)
sales_comparison = [kristin_metrics[0], tianna_metrics[0]]
plt.bar(['Kristin', 'Tianna'], sales_comparison)
plt.title('Total Sales Comparison')
plt.ylabel('Total Sales ($)')

# Transaction count comparison
plt.subplot(2, 2, 2)
transaction_comparison = [kristin_metrics[1], tianna_metrics[1]]
plt.bar(['Kristin', 'Tianna'], transaction_comparison)
plt.title('Number of Transactions Comparison')
plt.ylabel('Number of Transactions')

# Average transaction value comparison
plt.subplot(2, 2, 3)
avg_transaction_comparison = [kristin_metrics[2], tianna_metrics[2]]
plt.bar(['Kristin', 'Tianna'], avg_transaction_comparison)
plt.title('Average Transaction Value Comparison')
plt.ylabel('Average Transaction Value ($)')

# Product mix comparison
plt.subplot(2, 2, 4)
kristin_product_mix = kristin_metrics[3]
tianna_product_mix = tianna_metrics[3]
all_categories = list(set(kristin_product_mix.index) | set(tianna_product_mix.index))
kristin_values = [kristin_product_mix.get(cat, 0) for cat in all_categories]
tianna_values = [tianna_product_mix.get(cat, 0) for cat in all_categories]

x = range(len(all_categories))
width = 0.35
plt.bar([i - width/2 for i in x], kristin_values, width, label='Kristin')
plt.bar([i + width/2 for i in x], tianna_values, width, label='Tianna')
plt.xlabel('Product Categories')
plt.ylabel('Number of Transactions')
plt.title('Product Mix Comparison')
plt.xticks(x, all_categories, rotation=45, ha='right')
plt.legend()

plt.tight_layout()
plt.show()

# Sales over time
plt.figure(figsize=(12, 6))
kristin_daily_sales = kristin_data_matched.groupby('date_time_local')['sold_at_price'].sum()
tianna_daily_sales = tianna_data_matched.groupby('date_time_local')['sold_at_price'].sum()

plt.plot(kristin_daily_sales.index, kristin_daily_sales.values, label='Kristin')
plt.plot(tianna_daily_sales.index, tianna_daily_sales.values, label='Tianna')
plt.title('Daily Sales Comparison')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Cumulative sales over time
plt.figure(figsize=(12, 6))
kristin_cumulative_sales = kristin_daily_sales.cumsum()
tianna_cumulative_sales = tianna_daily_sales.cumsum()

plt.plot(kristin_cumulative_sales.index, kristin_cumulative_sales.values, label='Kristin')
plt.plot(tianna_cumulative_sales.index, tianna_cumulative_sales.values, label='Tianna')
plt.title('Cumulative Sales Comparison')
plt.xlabel('Date')
plt.ylabel('Cumulative Sales ($)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
